```python
# objetivo 2: lograr cargar todas las slices 

import pydicom
import matplotlib.pyplot as plt
import glob
import os

# Define la carpeta que contiene los archivos DICOM
folder_path = "D:/Facultad/5. Quinto ao/PRIMER SEMESTRE/PPS/ARCHIVOS DICOM/001. ABAD-36466-20230628T153733Z-001/001. ABAD-36466"

# Obtiene una lista de archivos DICOM que comienzan con "CT" en la carpeta
file_list = glob.glob(os.path.join(folder_path, "CT*.dcm"))

# Crea una lista para almacenar los objetos DICOM
slices = []

# Lee y almacena los archivos DICOM en la lista
for file in file_list:
    ds = pydicom.dcmread(file)
    slices.append(ds)

# Ordena los cortes segn SliceLocation
slices.sort(key=lambda x: float(x.SliceLocation))

# Ajusta el tamao de la figura
plt.figure(figsize=(300, 280))

# Itera a travs de los cortes y los muestra uno debajo del otro
for i, ds in enumerate(slices):
    plt.subplot(len(slices), 1, i + 1)
    plt.imshow(ds.pixel_array, cmap='gray')
    plt.title(f"Slice {i + 1}")
    plt.axis('off')

plt.show()
```
